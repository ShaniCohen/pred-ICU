```python
data_file_path = '../Data/training_v2.csv.zip'
```

```python
data_df = pd.read_csv(data_file_path)
data_df.head()
```

```python
print(f'The dataset has [{len(data_df.columns)}] columns and [{len(data_df)}] records')
```

```python
print(f'The dataset has [{len(data_df.patient_id.unique())}] patients (no multiple entries for patients).')
```

```python
data_no_apache_df = data_df.drop(list(data_df.filter(regex='apache')), axis=1)
print(f'The dataset has [{len(data_no_apache_df.columns)}] columns and [{len(data_no_apache_df)}] records (without apache scores)')
```

```python
survival_df = data_no_apache_df.hospital_death.value_counts()
colors = sns.color_palette('deep')[0:2]
plt.pie(survival_df, labels=['Survival', 'Mortality'], colors=colors, startangle=90, autopct='%1.1f%%')
plt.show()
```

```python
gender_df = data_no_apache_df['gender'].value_counts()
colors = sns.color_palette('deep')[0:2]
plt.pie(gender_df, labels=['Male', 'Female'], colors=colors, startangle=90, autopct='%1.1f%%')
plt.show()
```

```python
ethnicity_df = data_no_apache_df['ethnicity'].value_counts()
colors = sns.color_palette('deep')[0:6]
plt.pie(ethnicity_df, colors=colors, startangle=90)
labels = [f'{ethnicity} ({round(count * 100 / sum(ethnicity_df), 2)}%)' for ethnicity, count in zip(ethnicity_df.keys(), ethnicity_df)]
plt.legend(labels=labels, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
age_and_mortality_df = data_no_apache_df[['age', 'hospital_death']].groupby('age', as_index=False).value_counts()

age_and_mortality_df = pd.DataFrame(data_no_apache_df[['age', 'hospital_death']])

age_and_mortality_df['age_group'] = pd.cut(age_and_mortality_df['age'], bins=range(0, 110, 10), right=False)

ratio_df = age_and_mortality_df.groupby('age_group', observed=True)['hospital_death'].value_counts(normalize=True).unstack().fillna(0)
ratio_df['death_to_total_ratio'] = ratio_df[1] / (ratio_df[0] + ratio_df[1])

plt.bar(ratio_df.index.astype(str), ratio_df['death_to_total_ratio'])
plt.xlabel('Age Group (in 10-year intervals)')
plt.ylabel('Death to Total Ratio')
plt.title('Death to Total Ratio by Age Group')
plt.show()
```
