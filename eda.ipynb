```python
data_file_path = '../Data/training_v2.csv.zip'
```

```python
data_df = pd.read_csv(data_file_path)
data_df.head()
```

```python
print(f'The dataset has [{len(data_df.columns)}] columns and [{len(data_df)}] records')
```

```python
print(f'The dataset has [{len(data_df.patient_id.unique())}] patients (no multiple entries for patients).')
```

```python
data_no_apache_df = data_df.drop(list(data_df.filter(regex='apache')), axis=1)
print(f'The dataset has [{len(data_no_apache_df.columns)}] columns and [{len(data_no_apache_df)}] records (without apache scores)')
```

```python
survival_df = data_no_apache_df.hospital_death.value_counts()
colors = sns.color_palette('deep')[0:2]
plt.pie(survival_df, labels=['Survival', 'Mortality'], colors=colors, startangle=90, autopct='%1.1f%%')
plt.title(f'Mortality Rate (N={len(data_df.patient_id.unique())})')
plt.show()
```

```python
gender_df = data_no_apache_df['gender'].value_counts()
colors = sns.color_palette('deep')[0:2]
plt.pie(gender_df, labels=['Male', 'Female'], colors=colors, startangle=90, autopct='%1.1f%%')
plt.title(f'Gender Distribution (N={len(data_df.patient_id.unique())})')
plt.show()
```

```python
data_no_apache_filled_ethnicity_na_df = data_no_apache_df['ethnicity']
data_no_apache_filled_ethnicity_na_df[data_no_apache_filled_ethnicity_na_df.isna()] = 'Other/Unknown'
ethnicity_df = data_no_apache_filled_ethnicity_na_df.value_counts()
colors = sns.color_palette('deep')[0:6]
plt.pie(ethnicity_df, colors=colors, startangle=90)
labels = [f'{ethnicity} ({round(count * 100 / sum(ethnicity_df), 2)}%)' for ethnicity, count in zip(ethnicity_df.keys(), ethnicity_df)]
plt.legend(labels=labels, loc='center left', bbox_to_anchor=(1, 0.5))
# plt.title(f'Ethnicity Distribution (N={len(data_df.patient_id.unique())})')
plt.show()
```

```python
age_df = data_no_apache_df['age']
bins=74
plt.hist(age_df, bins=bins)
plt.title(f'Age Distribution')
plt.show()
```

```python
age_and_mortality_df = data_no_apache_df[['age', 'hospital_death']].groupby('age', as_index=False).value_counts()

age_and_mortality_df = pd.DataFrame(data_no_apache_df[['age', 'hospital_death']])

age_and_mortality_df['age_group'] = pd.cut(age_and_mortality_df['age'], bins=range(0, 110, 10), right=False)

ratio_df = age_and_mortality_df.groupby('age_group', observed=True)['hospital_death'].value_counts(normalize=True).unstack().fillna(0)
ratio_df['death_to_total_ratio'] = ratio_df[1] * 100 / (ratio_df[0] + ratio_df[1])

plt.bar(ratio_df.index.astype(str), ratio_df['death_to_total_ratio'])
plt.xlabel('Age Group (in 10-year intervals)')
plt.ylabel('Mortality percentage (%)')
plt.title(f'Mortality Percentage by Age Group (N={len(data_df.patient_id.unique())})')
plt.show()
```

```python
pre_icu_los_days_neg_count = data_df[data_df['pre_icu_los_days'] < 0].shape[0]
pre_icu_los_days_neg_percentage = round(pre_icu_los_days_neg_count * 100 / data_df.shape[0], 2)
print(f'pre_icu_los_days negative values: count - {pre_icu_los_days_neg_count}, percentage - {pre_icu_los_days_neg_percentage}%')
```

```python
pre_icu_los_days_pos_count = data_df[data_df['pre_icu_los_days'] >= 0].shape[0]
pre_icu_los_days_pos_percentage = round(pre_icu_los_days_pos_count * 100 / data_df.shape[0], 2)
print(f'pre_icu_los_days positive values: count - {pre_icu_los_days_pos_count}, percentage - {pre_icu_los_days_pos_percentage}%')
```

```python
# categorial values, both strings and numeric values
categorial_values = [
    'hospital_id',
    'icu_id',
    'gender',
    'elective_surgery',
    'ethnicity',
    'hospital_admit_source',
    'icu_admit_source',
    'icu_stay_type',
    'icu_type',
    'apache_post_operative',
    'arf_apache',
    'gcs_eyes_apache',
    'gcs_motor_apache',
    'gcs_unable_apache',
    'gcs_verbal_apache',
    'intubated_apache',
    'ventilated_apache',
    'aids',
    'cirrhosis',
    'diabetes_mellitus',
    'hepatic_failure',
    'immunosuppression',
    'leukemia',
    'lymphoma',
    'solid_tumor_with_metastasis',
    'apache_3j_bodysystem',
    'apache_2_bodysystem'
]

categorial_df = data_df[categorial_values]
categorial_df.nunique().sort_values(ascending=False).head(10)

categorial_unique_values_count_df = categorial_df.nunique().sort_values(ascending=False).head(10)

categorial_unique_values_count_df = categorial_unique_values_count_df.reset_index().rename(columns={'index': 'feature', 0: 'unique values count'})
categorial_unique_values_count_df.plot.bar(x='feature')
plt.title('Top-10 Categorial Features Unique Value Count')
plt.xlabel('')
plt.ylabel('# unique values')
```

```python
identifiers_columns = ['encounter_id', 'patient_id']
target_column = ['hospital_death']
apache_prediction_columns = ['apache_4a_hospital_death_prob', 'apache_4a_icu_death_prob']
const_values_columns = ['readmission_status']

non_categorial_df = data_df.drop(
    categorial_values + identifiers_columns + target_column + apache_prediction_columns,
    axis=1
)
non_categorial_unique_values_count_df = non_categorial_df.nunique().sort_values(ascending=False).head(10)

non_categorial_unique_values_count_df = non_categorial_unique_values_count_df.reset_index().rename(columns={'index': 'feature', 0: 'unique values count'})
non_categorial_unique_values_count_df.plot.bar(x='feature')
plt.title('Top-10 Non-Categorial Features Unique Value Count')
plt.xlabel('')
plt.ylabel('# unique values')
```

```python
map_apache_df = data_df['map_apache']
map_apache_non_integer_values = map_apache_df[map_apache_df.apply(lambda x: not x.is_integer())]
map_apache_null_values = map_apache_df.isnull()
# (map_apache_null_values & map_apache_non_integer_values).sum()

print(f'Existing map_apache values: [{map_apache_df[~map_apache_null_values].shape[0]}]')
```

```python
from sklearn.cluster import KMeans
# map_apache_hist = np.histogram(map_apache_df[~map_apache_null_values], bins=50)
# map_apache_hist = np.vstack((0.5*(map_apache_hist[1][:-1] + map_apache_hist[1][1:]), map_apache_hist[0])).T

map_apache_nparray = np.array(map_apache_df[~map_apache_null_values])

map_apache_kmeans = KMeans(n_clusters=2, random_state=1).fit(map_apache_nparray.reshape(map_apache_df[~map_apache_null_values].shape[0], 1))
# print(kmeans.cluster_centers_)
# kmeans.predict(np.array(93).reshape(-1 ,1))
# print(np.bincount(kmeans.labels_))
map_apache_kmeans_bin_count = np.bincount(map_apache_kmeans.labels_)
print(f'Cluster 1: center: {map_apache_kmeans.cluster_centers_[0]}, count: [{map_apache_kmeans_bin_count[0]}]')
print(f'Cluster 2: center: {map_apache_kmeans.cluster_centers_[1]}, count: [{map_apache_kmeans_bin_count[1]}]')
print(f'Total count: [{map_apache_nparray.shape[0]}]')
```

```python
heart_rate_apache_df = data_df['heart_rate_apache']

heart_rate_apache_null_values = heart_rate_apache_df.isna()
heart_rate_apache_nparray = np.array(heart_rate_apache_df[~heart_rate_apache_null_values])

heart_rate_apache_kmeans = KMeans(n_clusters=2, random_state=1).fit(heart_rate_apache_nparray.reshape(heart_rate_apache_df[~heart_rate_apache_null_values].shape[0], 1))
# # # print(kmeans.cluster_centers_)
# # # kmeans.predict(np.array(93).reshape(-1 ,1))
# # # print(np.bincount(kmeans.labels_))
heart_rate_apache_kmeans_bin_count = np.bincount(heart_rate_apache_kmeans.labels_)
print(f'Cluster 1: center: {heart_rate_apache_kmeans.cluster_centers_[0]}, count: [{heart_rate_apache_kmeans_bin_count[0]}]')
print(f'Cluster 2: center: {heart_rate_apache_kmeans.cluster_centers_[1]}, count: [{heart_rate_apache_kmeans_bin_count[1]}]')
print(f'Total count: [{heart_rate_apache_nparray.shape[0]}]')
```
